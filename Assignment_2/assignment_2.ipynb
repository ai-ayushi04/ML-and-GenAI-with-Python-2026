```python
#Student Result System (with loops + if-elif-else)
name = input("Enter student name: ")
roll = input("Enter roll number: ")
subjects = int(input("Enter number of subjects: "))
total = 0
for i in range(subjects):
    marks = float(input(f"Enter marks of subject {i+1}: "))
    total += marks
percentage = (total / (subjects * 100)) * 100
print("\n----- RESULT -----")
print("Name:", name)
print("Roll No:", roll)
print("Total Marks:", total)
print("Percentage:", percentage)
if percentage >= 75:
    print("Grade: A")
elif percentage >= 60:
    print("Grade: B")
elif percentage >= 40:
    print("Grade: C")
else:
    print("Grade: F")
```
